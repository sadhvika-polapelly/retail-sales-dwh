```python

%sql
USE CATALOG retail_dwh;
USE DATABASE silver;
```

```python
%sql
SELECT 'Bronze Customers'    AS layer_table, COUNT(*) AS row_count FROM retail_dwh.bronze.customers
UNION ALL
SELECT 'Silver Customers',    COUNT(*) FROM retail_dwh.silver.customers
UNION ALL
SELECT 'Bronze Products',     COUNT(*) FROM retail_dwh.bronze.products
UNION ALL
SELECT 'Silver Products',     COUNT(*) FROM retail_dwh.silver.products
UNION ALL
SELECT 'Bronze Stores',       COUNT(*) FROM retail_dwh.bronze.stores
UNION ALL
SELECT 'Silver Stores',       COUNT(*) FROM retail_dwh.silver.stores
UNION ALL
SELECT 'Bronze Sales',        COUNT(*) FROM retail_dwh.bronze.sales_transactions
UNION ALL
SELECT 'Silver Sales',        COUNT(*) FROM retail_dwh.silver.sales_transactions;
```

```python
%sql
SELECT CustomerID, COUNT(*) as duplicate_count
FROM retail_dwh.silver.customers
GROUP BY CustomerID
HAVING COUNT(*) > 1;
```

```python
%sql
SELECT ProductID, COUNT(*) AS duplicate_count
FROM retail_dwh.silver.products
GROUP BY ProductID
HAVING COUNT(*) > 1;
```

```python
%sql
SELECT COUNT(*) AS null_productid_count
FROM retail_dwh.silver.products
WHERE ProductID IS NULL;
```

```python
%sql
SELECT COUNT(*) AS invalidemail_count
FROM retail_dwh.silver.customers
WHERE Email NOT LIKE '%@%.%';
```

```python
%sql
--uppercase email check--
SELECT COUNT(*) AS uppercase_email_count
FROM retail_dwh.silver.customers
WHERE Email != LOWER(Email);
```

```python
%sql
--unitprice check--
SELECT COUNT(*) AS zero_price_count
FROM retail_dwh.silver.products
WHERE UnitPrice = 0;
```

```python
%sql
SELECT COUNT(*) AS null_region_count
FROM retail_dwh.silver.stores
WHERE Region IS NULL;
```

```python
%sql
SELECT COUNT(*)
FROM retail_dwh.silver.sales_transactions
WHERE Quantity <= 0;
```

```python
%sql
SELECT COUNT(*) as future_transactiondate
FROM retail_dwh.silver.sales_transactions
WHERE TxnDate > CURRENT_DATE();
```

```python
%sql

SELECT COUNT(*) AS city_spaces_count
FROM retail_dwh.silver.customers
WHERE City != TRIM(City);
```

```python
%sql

SELECT COUNT(*) AS productname_spaces_count
FROM retail_dwh.silver.products
WHERE ProductName != TRIM(ProductName);
```

```python
%sql

SELECT COUNT(*) AS inconsistent_storecase_count
FROM retail_dwh.silver.stores
WHERE StoreName != INITCAP(LOWER(StoreName));
```

```python
%sql

SELECT COUNT(*) AS invalid_customer_reference_count
FROM (
    SELECT s.CustomerID
    FROM retail_dwh.silver.sales_transactions s
    LEFT JOIN retail_dwh.silver.customers c
    ON s.CustomerID = c.CustomerID
    WHERE c.CustomerID IS NULL
);
```
