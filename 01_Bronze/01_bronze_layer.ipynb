```python
%sql
USE CATALOG retail_dwh;
USE DATABASE bronze;
```

```python
%sql
DROP TABLE IF EXISTS retail_dwh.bronze.customers;
DROP TABLE IF EXISTS retail_dwh.bronze.products;
DROP TABLE IF EXISTS retail_dwh.bronze.stores;
DROP TABLE IF EXISTS retail_dwh.bronze.sales_transactions;
```

```python
%sql
CREATE OR REPLACE TABLE retail_dwh.bronze.customers
USING DELTA
LOCATION 's3a://retail-sales-dwh-sadhvika/bronze/customers'
AS
SELECT 
    CustomerID,
    CustomerName,
    Email,
    City,
    Address,
    LastUpdated,
    current_timestamp() AS ingestion_timestamp,
    'customers_src_*.csv' AS source_file
FROM read_files(
    's3a://retail-sales-dwh-sadhvika/sftp/customers_src_*.csv',
    format => 'csv',
    header => true,
    inferSchema => true
);
```

```python
%sql
    
CREATE OR REPLACE TABLE retail_dwh.bronze.products
USING DELTA
LOCATION 's3a://retail-sales-dwh-sadhvika/bronze/products'
AS
SELECT
    ProductID,
    ProductName,
    Category,
    UnitPrice,
    current_timestamp() AS ingestion_timestamp,
    'products_src_*.csv' AS source_file
FROM read_files(
    's3a://retail-sales-dwh-sadhvika/sftp/products_src_*.csv',
    format => 'csv',
    header => true,
    inferSchema => true
);
```

```python
%sql
    
CREATE OR REPLACE TABLE retail_dwh.bronze.stores
USING DELTA
LOCATION 's3a://retail-sales-dwh-sadhvika/bronze/stores'
AS
SELECT
    StoreID,
    StoreName,
    Region,
    current_timestamp() AS ingestion_timestamp,
    'stores_src_*.csv' AS source_file
FROM read_files(
    's3a://retail-sales-dwh-sadhvika/sftp/stores_src_*.csv',
    format => 'csv',
    header => true,
    inferSchema => true
);
```

```python
%sql
    
CREATE OR REPLACE TABLE retail_dwh.bronze.sales_transactions
USING DELTA
LOCATION 's3a://retail-sales-dwh-sadhvika/bronze/sales_transactions'
AS
SELECT
    TransactionID,
    CustomerID,
    ProductID,
    StoreID,
    Quantity,
    TxnDate,
    current_timestamp() AS ingestion_timestamp,
    'sales_transactions_src_*.csv' AS source_file
FROM read_files(
    's3a://retail-sales-dwh-sadhvika/sftp/sales_transactions_src_*.csv',
    format => 'csv',
    header => true,
    inferSchema => true
);
```

```python
%sql
SELECT 'customers' AS table_name, COUNT(*) AS row_count FROM retail_dwh.bronze.customers
UNION ALL
SELECT 'products', COUNT(*) FROM retail_dwh.bronze.products
UNION ALL
SELECT 'stores', COUNT(*) FROM retail_dwh.bronze.stores
UNION ALL
SELECT 'sales_transactions', COUNT(*) FROM retail_dwh.bronze.sales_transactions;
```
